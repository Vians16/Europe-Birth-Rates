average consumer prices Annual percent change

```python
#Import Library
import pandas as pd
```

```python
#Import CSV
df= pd.read_csv(r'C:\Users\Admin\Desktop\GOOGLE DATA CERTIFICATE\CAPSTONE_GOOGLE_CERT\database_europe birthrate\DATABASE\europe inflation\Inflation_rates_worldwide.csv')
df.head(2)
```

```python
#Drop first row because is empty
df.drop(index=0, inplace=True)
df.head(2)
```

```python
#Rename Inflation rate, average consumer prices (Annual percent change) column
df.rename(columns={'Inflation rate, average consumer prices (Annual percent change)': 'Inflation Rate'}, inplace=True)
df.head(2)

```

```python
#Filter the years between 2000 & 2023
columns_to_keep= ['Inflation Rate']+[col for col in df.columns if col.isdigit() and 2000 <= int(col) <=2023]
df = df[columns_to_keep]
df.head(2)
```

```python
#Rename all no data values with a blank
df.replace('no data','',inplace=True)
df.head(2)
```

```python
#Export Excel 
Output_Path= (r'C:\Users\Admin\Desktop\GOOGLE DATA CERTIFICATE\CAPSTONE_GOOGLE_CERT\database_europe birthrate\DATABASE\europe inflation\Inflation_worldwide.xlsx')
df.to_excel(Output_Path)
```
