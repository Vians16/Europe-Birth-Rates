```python
import pandas as pd
```

```python
df= pd.read_csv(r'C:\Users\Admin\Desktop\GOOGLE DATA CERTIFICATE\CAPSTONE_GOOGLE_CERT\database_europe birthrate\DATABASE\europe population\european-union-TOTAL-population.csv')
df.head(3)
```

```python
#Drop Indicator Name & Indicator Code columns
drop_cols=['Indicator Code',]
df = df.drop(columns=drop_cols)
df.head(3)
```

```python
#Drop all columns where year < 2000's
columns_to_keep = ['Country Name'] + ['Indicator Name'] + ['Country Code'] + [col for col in df.columns if col.isdigit() and int(col) >= 2000]
df = df[columns_to_keep]
df.head(3)
```

```python
#Reorder the columns
Col_order =['Country Name','Country Code','Indicator Name']+[col for col in df.columns if col.isdigit()]
df = df[Col_order]
```

```python
#Million format to the year columns
years_col=[col for col in df.columns if col.isdigit()]
df[years_col]= df[years_col].applymap(lambda x: f"{x:,.0f}" if pd.notnull(x) else x)
df.head(3)
```

```python
#Export to an Excel 
output_path= r'C:\Users\Admin\Desktop\GOOGLE DATA CERTIFICATE\CAPSTONE_GOOGLE_CERT\database_europe birthrate\DATABASE\europe population\TOTALPopulation_CLEAN.xlsx'
df.to_excel(output_path,index=False)
```
