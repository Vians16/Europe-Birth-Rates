```python
import pandas as pd
```

```python
#Read CSV file

df= pd.read_csv(r'C:\Users\Admin\Desktop\GOOGLE DATA CERTIFICATE\CAPSTONE_GOOGLE_CERT\database_europe birthrate\DATABASE\GDP around theworld\GDP.csv')
df.head(1)
```

```python
#Filtering the columns need it

df = df.filter(items=['Country Name', 'Country Code']).join(df.filter(regex=r'^20[0-9]{2}$'))
df.head(5)
```

```python
# Identify columns with year values (2000 and onwards)
year_columns = df.columns[df.columns.str.match(r'^20[0-9]{2}$')]

# Round the GDP values in those columns
df[year_columns] = df[year_columns].round(2)

# Apply formatting to add commas
df[year_columns] = df[year_columns].applymap(lambda x: f"{x:,.2f}")

# Display the first 5 rows to verify
df.head(5)
```

```python
#Export to excel
output_path= r'C:\Users\Admin\Desktop\GOOGLE DATA CERTIFICATE\CAPSTONE_GOOGLE_CERT\database_europe birthrate\DATABASE\GDP around theworld\GDP_Per_Capita_CLEAN.xlsx'
df= df.to_excel(output_path)
```
